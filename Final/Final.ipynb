```julia
#Q6
t = collect(range(-0.75, 0.25, length = 7));
A = [1 t[1] t[1]^2 t[1]^3 0 0 0 0; 1 t[2] t[2]^2 t[2]^3 0 0 0 0; 1 t[3] t[3]^2 t[3]^3 0 0 0 0; 0 0 0 0 1 t[5] t[5]^2 t[5]^3; 0 0 0 0 1 t[6] t[6]^2 t[6]^3; 0 0 0 0 1 t[7] t[7]^2 t[7]^3; 1 t[4] t[4]^2 t[4]^3 -1 -t[4] -t[4]^2 -t[4]^3; 0 1 2*t[4] 3*t[4]^2 0 -1 -2*t[4] -3*t[4]^2]
b = [0;-0.1;0.5;1;0.8;0.5;0;0];
x = A\b;
display(x);
p(t) = (x[1] + x[2]*t + x[3]*t^2 + x[4]*t^3);
q(t) = (x[5] + x[6]*t + x[7]*t^2 + x[8]*t^3);
plot(p, -0.75, 0.25,label = "p(t)",xlabel="t",ylabel="poly(t)")
plot!(q, -0.75, 0.25,label = "q(t)",xlabel="t",ylabel="poly(t)")
```
