```julia
open("Cluster_Data.txt", "w") do file
    for i in 1:10
        topic = "Topic Number: "*string(i)
        titles = article_titles[labels_2 .== i]
        words = dictionary[sortperm(centroids_2[i],rev=true)]
        ans = topic*"\nTitles: "*join(titles,",")*" \nCommon Words: "*join(words,",")*"\n\n"
        write(file, ans)
    end
end
```
